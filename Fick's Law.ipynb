```r
dL = 0.1 
dR = 0.1 
N0 = 2000 
x0 = 6 # initial placemanet of population 
nx = 11 #number of cells 

cells = vector()
for (z in 1:11) {

    cells[z] = 0
    if (z == 6) {
    
        cells[z] = N0
        
    }
    
}
data_matrix <- data.frame (
    "Time (h)" = 1, "Cell1" = cells[1],  "Cell2" = cells[2], "Cell3" = cells[3], "Cell4" = cells[4], "Cell5" = cells[5], "Cell6" = cells[6],
    "Cell7" = cells[7], "Cell8" = cells[8], "Cell9" = cells[9], "Cell10" = cells[10], "Cell11" = cells[11]
)
expected_matrix_1 <- data.frame (
    "Time (h)" = 1, "Cell1" = cells[1],  "Cell2" = cells[2], "Cell3" = cells[3], "Cell4" = cells[4], "Cell5" = cells[5], "Cell6" = cells[6],
    "Cell7" = cells[7], "Cell8" = cells[8], "Cell9" = cells[9], "Cell10" = cells[10], "Cell11" = cells[11]
)
expected_matrix_2 <- data.frame (
    "Time (h)" = 1, "Cell1" = cells[1],  "Cell2" = cells[2], "Cell3" = cells[3], "Cell4" = cells[4], "Cell5" = cells[5], "Cell6" = cells[6],
    "Cell7" = cells[7], "Cell8" = cells[8], "Cell9" = cells[9], "Cell10" = cells[10], "Cell11" = cells[11]
)
expected_matrix_3 <- data.frame (
    "Time (h)" = 1, "Cell1" = cells[1],  "Cell2" = cells[2], "Cell3" = cells[3], "Cell4" = cells[4], "Cell5" = cells[5], "Cell6" = cells[6],
    "Cell7" = cells[7], "Cell8" = cells[8], "Cell9" = cells[9], "Cell10" = cells[10], "Cell11" = cells[11]
)
```

```r
h = 90
hcells = vector() 
for (y in 1:h) {
     
    for (z in 2:10) { 
        
        hcells[z] =  cells[z] + (cells[z-1] * dR) + (cells[z+1] * dL) -  (cells[z] * (dL + dR))

    }
    hcells[1] = cells[1] + (cells[2] * dL) -  (cells[1] * (dR)) 
    hcells[11] = cells[11] + (cells[10] * dR) -  (cells[11] * (dL))

    for ( z in 1:11) {

        cells[z] = hcells[z] 
        
    } 
    x <- c ( y+1 , cells) 
    data_matrix <- rbind(data_matrix, x) 
}
```

```r
distance = vector() 
for ( z in 1:11) {

    distance[z] = (z - x0) ^ 2
    
}
```

```r
meandistance = vector()  

for ( y in 1:(h+1)) {

    meandistance[y] = 0 
    for (z in 1:11) {
        
        meandistance[y] = meandistance[y] + distance[z] * data_matrix [y , z+1] 
            
    }
    meandistance[y] = meandistance[y] / N0
    
}
```

```r
plot (0:(h) , meandistance , ylab = "Variance", xlab = "Time", type = "l")
```

```r
D = vector() 

for ( y in 1:(h+1) ) {

    D[y]  = meandistance[y] / (2 * (y-1)) 
    
}
```

```r
plot (0:(h) , D , ylab = "D", xlab = "Time", type = "l")
```

```r
df <- data.frame("Variance" = meandistance, "Time" = 0:h )

df.lm <- lm(Variance ~ Time, data = df)

summary(df.lm)
```
